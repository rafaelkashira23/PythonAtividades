```python
tempo = float(input("Informe o tempo gasto em horas: "))
velocidade = float(input("Informe a velocidade média (em km/h): "))
distancia = velocidade * tempo
consumoMedio = float(input('Qaul o consumo médio de combustível (Km/L): '))
def cons(distancia,consumoMedio):
  consumo = distancia/consumoMedio
  return consumo
cons(distancia,consumoMedio)
```
