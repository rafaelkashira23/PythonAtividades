```python
#carregando train.csv para um dataframe

import pandas as pd

df = pd.read_csv('train.csv')
print(df.head())
```

```python
# removendo as linhas com valores faltantes

df = df.dropna()
print(df.head())
```

```python
# Gerando os gráficos:
# - Distribuição da classe de passageiros (PClass) por categoria de sobreviventes (Survived - sobreviventes e não sobreviventes);
# - Distribuição de sexo dos passageiros (Sex) por categoria de sobreviventes (Survived - sobreviventes e não sobreviventes);

import matplotlib.pyplot as plt
import seaborn as sns

# Distribuição da classe de passageiros (PClass) por categoria de sobreviventes
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Distribuição da Classe de Passageiros por Sobrevivência')
plt.xlabel('Classe de Passageiros')
plt.ylabel('Contagem')
plt.show()

# Distribuição de sexo dos passageiros (Sex) por categoria de sobreviventes
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Distribuição de Sexo por Sobrevivência')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.show()
```
